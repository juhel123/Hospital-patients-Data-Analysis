How many patients are there in the database?

```python
Number_of_patients = data.groupby('FIRST')['LAST'].value_counts()
Number_of_patients
```

```python
plt.figure(figsize=(12,6))
Number_of_patients.plot(kind='bar')
plt.xlabel('Name of patients')
plt.ylabel('Count of patients')
plt.title('Total number of Patients')
plt.xticks(rotation=45, fontsize=2)
plt.show()
```

Top two cities with the highest birthday?

```python
# Group by city and count the number of birthdays (or people)
city_counts = data.groupby('CITY').size().reset_index(name='No of births')

# Sort by birthday_count descending and get top 2
top_two_cities = city_counts.sort_values(by='No of births', ascending=False).head(2)

print(top_two_cities)
```

```python
# Plot the top two cities
top_two_cities.plot(kind='bar', color=['blue','red'])
plt.title('Top Two Cities with Highest Birthdays')
plt.xlabel('City')
plt.ylabel('Total Birthday Count')
plt.xticks(rotation=45)
plt.show()
```

 Top two cities with the highest deathdate?

```python
# Assuming df has columns 'City' and 'deathdate'
# Count number of deaths per city
death_counts = data.groupby('CITY')['DEATHDATE'].count()

# Sort counts descending and get top two
top_two_cities = death_counts.sort_values(ascending=False).head(2)

print(top_two_cities)
```

```python
# Plot the top two cities
top_two_cities.plot(kind='bar', color=['blue', 'green'])
plt.title('Top Two Cities with Highest Birthdays')
plt.xlabel('City')
plt.ylabel('Total Birthday Count')
plt.xticks(rotation=270)
plt.show()
```
